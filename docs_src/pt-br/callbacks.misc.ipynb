Poderia haver vários usos para este callback calhar.
O objetivo inicial era para ser capaz de verificar rapidamente os requisitos de memória para um determinado conjunto de hyperparamaters como `bs` e` size`.
Desde toda a memória GPU necessário é configurado durante o primeiro lote do primeiro [see tutorial](https://docs.fast.ai/tutorial.resources.html#gpu-memory-usage-anatomy) época, é o suficiente para rodar apenas 1-2 lotes para medir se os hiperparâmetros está certo e não vai levar a erros (OOM) fora-de-memória. Então, ao invés de esperar por minutos ou horas para apenas descobrir que seu `bs` ou` size` são muito grandes, este callback permite que você faça isso segundos.
Você pode implantá-lo em um aluno específico (ou chamada de ajuste), assim como com qualquer outra chamada de retorno:
```
from fastai.callbacks.misc import StopAfterNBatches
[...]
learn = cnn_learner([...])
learn.callbacks.append(StopAfterNBatches(learn, n_batches=2))
learn.fit_one_cycle(3, max_lr=1e-2)
```
e ele vai quer caber na memória existente ou ele vai falhar imediatamente com o erro OOM. Você pode querer adicionar [ipyexperiments](https://github.com/stas00/ipyexperiments/) para lhe mostrar o uso de memória, incluindo o pico de uso.
Isso é bom, mas é complicado, pois você tem que mudar o código fonte notebook e muitas vezes você vai ter vários alunos e chamadas cabem no mesmo notebook, então aqui está como fazê-lo globalmente, colocando o seguinte código em algum lugar no topo do seu notebook e deixando o resto do seu notebook não modificada:
```
from fastai.callbacks.misc import StopAfterNBatches
# True turns the speedup on, False return to normal behavior
tune = True
#tune = False
if tune:
    defaults.silent = True # don't print the fit metric headers
    defaults.extra_callbacks = [StopAfterNBatches(n_batches=2)]
else:
    defaults.silent = False
    defaults.extra_callbacks = None  
```
Quando você terminar de ajustar seus hiper-parâmetros, basta definir `tune` para` false` e ​​re-executar o notebook para fazer verdadeiro encaixe.
A configuração `defaults.silent` controla se as chamadas [`fit`](/basic_train.html#fit) imprimir qualquer saída.
Note que quando executa este retorno de chamada, cada chamada ajuste será interrompido resultando na saída de cor vermelha - que é apenas uma indicação de que o ajuste normal não acontecer, então você não deve esperar quaisquer resultados qualitativos fora dele.
