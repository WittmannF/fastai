## modelo simples para dados tabulares

```python
from fastai.gen_doc.nbdoc import *
from fastai.tabular.models import TabularModel
```

```python
show_doc(TabularModel)
```

`Emb_szs` corresponder cada tamanho variável categórica com um tamanho de incorporação,` n_cont` é o número de variáveis ​​contínuas. O modelo consiste em camadas `Embedding` para as variáveis ​​categóricas, seguido por um` `Dropout` de emb_drop`, e um` BatchNorm` para as variáveis ​​contínuas. Os resultados são concatenadas e seguido por blocos de `BatchNorm`,` Dropout`, `e` Linear` ReLU` (os primeiros salta bloco `` BatchNorm` e Dropout`, o último bloco ignora a `ReLU`).
Os tamanhos dos blocos são dadas em [`layers`](/layers.html#layers) e as probabilidades de a `` Dropout` em ps`. O último tamanho é `out_sz`, e nós adicionamos uma última ativação, que é um sigmóide redimensionado para cobrir` y_range` (se não é `none`). Por último, se `use_bn` é definido como False, todos os` camadas BatchNorm` são ignorados, exceto aquele aplicado às variáveis ​​contínuas.
Geralmente é mais fácil simplesmente criar um aluno com [`tabular_learner`](/tabular.learner.html#tabular_learner), que criará automaticamente um [`TabularModel`](/tabular.models.html#TabularModel) para você.
