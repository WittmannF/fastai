## Memória Tracing Utils

````python
show_doc(GPUMemTrace, title_level=4)
````

** Argumentos **:
* `Silent`: um atalho para fazer` report` e `report_n_reset` silencioso w / o a necessidade de remover as chamadas - isso pode ser feito a partir do construtor, ou, alternativamente, você pode chamar` método silent` em qualquer lugar para fazer o mesmo.
* `Ctx`: Nota de contexto padrão em relatórios
* `On_exit_report`: auto-relatório sobre saída gerente ctx (padrão` true`)
** Definições **:
* ** Delta Usado ** é a diferença entre a memória usada atual e memória usada no início do balcão.
* ** Delta repicado ** é a sobrecarga de memória, se houver. Ele é calculado em duas etapas:
1. A base é medida a diferença entre a memória de pico e a memória usada no início do contador.
2. Então, se delta utilizado é positivo ele é subtraído do valor base.

Ele indica o tamanho do blip.
** Atenção **: atualmente o rastreamento de pico de uso de memória é implementado usando um fio python, que é muito confiável, já que não há garantia do fio vai ter a chance de correr no momento em que a memória de pico está ocorrendo (ou talvez não ter a chance de correr em tudo). Portanto, precisamos pytorch para implementar vários contadores [`torch.cuda.max_memory_allocated`](https://pytorch.org/docs/stable/cuda.html#torch.cuda.max_memory_allocated) concorrentes e repostos. Por favor vote para este [feature request](https://github.com/pytorch/pytorch/issues/16266).
** ** Exemplos de uso:
Configuração:
```
from fastai.utils.mem import GPUMemTrace
def some_code(): pass
mtrace = GPUMemTrace()
```
Exemplo 1: Medições de base via `report` (gravuras) e via [`data`](/tabular.data.html#tabular.data) (retornos) assessores
```
some_code()
mtrace.report()
delta_used, delta_peaked = mtrace.data()

some_code()
mtrace.report('2nd run of some_code()')
delta_used, delta_peaked = mtrace.data()
```
`` Opcional argumento subctx` de report` pode ser útil se você tiver muitos `chamadas report` e você quer entender o que é que nas saídas.
Exemplo 2: medir em um loop, redefinindo o contador antes de cada execução
```
for i in range(10):
    mtrace.reset()
    some_code()
    mtrace.report(f'i={i}')
```
`Reset` redefine todos os contadores.
Exemplo 3: como o exemplo 2, mas tendo `report` repor automaticamente os contadores
```
mtrace.reset()
for i in range(10):
    some_code()
    mtrace.report_n_reset(f'i={i}')
```
O traçado começa imediatamente após a criação do objeto [`GPUMemTrace`](/utils.mem.html#GPUMemTrace), e pára quando esse objeto é excluído. Mas pode também ser `stop`ed,` start`ed manualmente.
```
mtrace.start()
mtrace.stop()
```
`Stop` é especialmente útil se você quiser ** congelamento ** o objeto [`GPUMemTrace`](/utils.mem.html#GPUMemTrace) e ser capaz de consultar seus dados em` stop` algum tempo na estrada.
**Comunicando**:
Nos relatórios, você pode imprimir um contexto principal passou através do construtor:
```
mtrace = GPUMemTrace(ctx="foobar")
mtrace.report()
```
impressões:
```
△Used Peaked MB:      0      0  (foobar)
```
e, em seguida, adicionar notas subcontexto, conforme necessário:
```
mtrace = GPUMemTrace(ctx="foobar")
mtrace.report('1st try')
mtrace.report('2nd try')

```
impressões:
```
△Used Peaked MB:      0      0  (foobar: 1st try)
△Used Peaked MB:      0      0  (foobar: 2nd try)
```
Ambos contexto e sub-contexto são opcionais, e são muito útil se você polvilhe [`GPUMemTrace`](/utils.mem.html#GPUMemTrace) em locais diferentes ao redor do código.
Você pode silenciar relatório apela w / o a necessidade de removê-los via construtor ou `silent`:
```
mtrace = GPUMemTrace(silent=True)
mtrace.report() # nothing will be printed
mtrace.silent(silent=False)
mtrace.report() # printing resumed
mtrace.silent(silent=True)
mtrace.report() # nothing will be printed
```
** Contexto Gerente **:
[`GPUMemTrace`](/utils.mem.html#GPUMemTrace) também pode ser usado como um gerente de contexto:
Relatar os deltas usados ​​e pontiagudos automaticamente:
```
with GPUMemTrace(): some_code()
```
Se você deseja adicionar contexto:
```
with GPUMemTrace(ctx='some context'): some_code()
```
O gerente de contexto usa subcontexto `exit` para indicar que o relatório vem após o contexto saiu.
O relato é feito automaticamente, o que é especialmente útil em funções devido ao retorno de chamada:
```
def some_func():
    with GPUMemTrace(ctx='some_func'):
        # some code
        return 1
some_func()
```
impressões:
```
△Used Peaked MB:      0      0 (some_func: exit)
```
assim você ainda obter um relatório perfeito, apesar do `chamada return` aqui. `Ctx` é útil para especificar o contexto * * no caso de você ter muitas dessas chamadas através de seu código e você quer saber qual é qual.
E, claro, em vez de fazer o acima, você pode usar [`gpu_mem_trace`](/utils.mem.html#gpu_mem_trace) decorador para fazer isso automaticamente, incluindo o uso da função ou nome do método como o contexto. Portanto, o exemplo que se segue faz a mesma sem modificar a função.
```
@gpu_mem_trace
def some_func():
    # some code
    return 1
some_func()
```
Se você não quiser que o relatório automático, basta passar `on_exit_report = false` no construtor:
```
with GPUMemTrace(ctx='some_func', on_exit_report=False) as mtrace:
    some_code()
mtrace.report("measured in ctx")
```
ou o mesmo w / o a nota de contexto:
```
with GPUMemTrace(on_exit_report=False) as mtrace: some_code()
print(mtrace) # or mtrace.report()
```
E, claro, você pode obter os dados numéricos (em MBs arredondados):
```
with GPUMemTrace() as mtrace: some_code()
delta_used, delta_peaked = mtrace.data()
```

````python
show_doc(gpu_mem_trace)
````

Isso permite que você decorar qualquer função ou método com:
```
@gpu_mem_trace
def my_function(): pass
# run:
my_function()
```
e ele irá imprimir automaticamente o relatório, incluindo o nome da função como um contexto:
```
△Used Peaked MB:      0      0 (my_function: exit)
```
No caso de métodos que vai imprimir um método totalmente qualificado, por exemplo .:
```
△Used Peaked MB:      0      0 (Class.function: exit)
```
