Aqui, novamente, a espinha dorsal do modelo é determinado por `arch` e` config`. Os textos de entrada são alimentados em que o modelo monte de `bptt` e só nos últimos` ativações max_len` são considerados. Isso nos dá a espinha dorsal do nosso modelo. A cabeça, em seguida, consiste em:
- uma camada que encadeia os produtos finais da RNN com o máximo e a média de todas as saídas intermédias (sobre a dimensão comprimento da sequência),
- os blocos de camadas ([`nn.BatchNorm1d`](https://pytorch.org/docs/stable/nn.html#torch.nn.BatchNorm1d), [`nn.Dropout`](https://pytorch.org/docs/stable/nn.html#torch.nn.Dropout), [`nn.Linear`](https://pytorch.org/docs/stable/nn.html#torch.nn.Linear), [`nn.ReLU`](https://pytorch.org/docs/stable/nn.html#torch.nn.ReLU)).
Os blocos são definidos pelo `e` lin_ftrs` argumentos drops`. Especificamente, o primeiro bloco terá um número de entradas inferidas a partir da espinha dorsal do arco e o último irá ter um número de saídas iguais data.c (que contém o número de classes de dados) e os blocos intermédios têm um número de entradas / saídas determinados por `lin_ftrs` (naturalmente um bloco tem uma série de entradas igual ao número de saídas do bloco anterior). As desistências têm um valor igual ps se você passar um float, ou os valores correspondentes se você passar uma lista. O padrão é ter um tamanho intermédio de oculto 50 (que faz com que dois blocos model_activation -> 50 -> n_classes) com uma saída de 0,1.

```python
show_doc(get_language_model)
```

Este modelo usa um codificador retirado do `arch` em` config`. Esse codificador é alimentado a sequência de bits sucessivos de tamanho bptt` `e só manter os últimos` saídas max_seq` para as camadas de agrupamento.
O decodificador usar uma concatenação das últimas saídas, um `MaxPooling` de todas as saídas e um` AveragePooling` de todas as saídas. Em seguida, utiliza uma lista de `BatchNorm`,` Dropout`, `Linear`,` blocos ReLU` (sem `ReLU` no último), utilizando um primeiro tamanho de camada` 3 * emb_sz` em seguida, seguindo os números entre `n_layers`. As probabilidades de desistências são lidos em `drops`.
Note-se que o modelo retorna uma lista de três coisas, a saída real de ser o primeiro, os outros dois sendo os estados ocultos intermediários antes e depois da desistência (usado pelo [`RNNTrainer`](/callbacks.rnn.html#RNNTrainer)). A maioria das funções de perda de esperar uma saída, então você deve usar uma chamada de retorno para remover os outros dois se você não estiver usando [`RNNTrainer`](/callbacks.rnn.html#RNNTrainer).
