## manipulação de dados tabulares

Este módulo define a classe principal para lidar com dados tabulares na biblioteca fastai: [`TabularDataBunch`](/tabular.data.html#TabularDataBunch). Como sempre, há também uma função auxiliar de obter rapidamente os seus dados.
Para permitir que você crie facilmente um [`Learner`](/basic_train.html#Learner) para os seus dados, ele fornece [`tabular_learner`](/tabular.learner.html#tabular_learner).

```python
from fastai.gen_doc.nbdoc import *
from fastai.tabular import * 
```

```python
show_doc(TabularDataBunch)
```

A melhor maneira de obter rapidamente os seus dados em um [`DataBunch`](/basic_data.html#DataBunch) adequado para dados tabulares é organizar-lo em duas (ou três) dataframes. Um para treinamento, um para validação, e se você tem isso, uma para testes. Aqui estamos interessados ​​em uma subamostra do [adult dataset](https://archive.ics.uci.edu/ml/datasets/adult).

```python
path = untar_data(URLs.ADULT_SAMPLE)
df = pd.read_csv(path/'adult.csv')
valid_idx = range(len(df)-2000, len(df))
df.head()
```

```python
cat_names = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
dep_var = 'salary'
```

A inicialização do [`TabularDataBunch`](/tabular.data.html#TabularDataBunch) é o mesmo que [`DataBunch`](/basic_data.html#DataBunch) para que você realmente quer usar o método de fábrica em seu lugar.

```python
show_doc(TabularDataBunch.from_df)
```

Opcionalmente, usar `test_df` para o conjunto de teste. A variável dependente é `dep_var`, enquanto as variáveis ​​categóricas e contínuas estão nas colunas` `cat_names` e colunas cont_names` respectivamente. Se `cont_names` é Nenhum, em seguida, assumimos todas as variáveis ​​que não são dependentes ou categóricas são contínuas. O [`TabularProcessor`](/tabular.data.html#TabularProcessor) em `procs` são aplicados aos dataframes como pré-processamento, em seguida, as categorias são substituídos pelos seus códigos + 1 (deixando para 0` nan`) e as variáveis ​​contínuas são normalizados.
Note que o [`TabularProcessor`](/tabular.data.html#TabularProcessor) deve ser passado como `Callable`: a inicialização real com` cat_names` e `cont_names` é feito durante o pré-processamento.

```python
procs = [FillMissing, Categorify, Normalize]
data = TabularDataBunch.from_df(path, df, dep_var, valid_idx=valid_idx, procs=procs, cat_names=cat_names)
```

Você pode facilmente criar um [`Learner`](/basic_train.html#Learner) para esses dados com [`tabular_learner`](/tabular.learner.html#tabular_learner).

```python
show_doc(tabular_learner)
```
