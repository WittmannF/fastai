## montar rapidamente seus dados

Você deve obter os seus dados em um dos seguintes formatos para aproveitar ao máximo a biblioteca fastai e utilizar um dos métodos de fábrica de uma das classes [`TextDataBunch`](/text.data.html#TextDataBunch):
- arquivos de texto cru em pastas trem, válido, teste em um estilo IMAGEnet,
- CSV onde alguns coluna (s) dá o rótulo (s) e o seguinte o texto associado,
- uma trama de dados estruturados da mesma forma,
- fichas e etiquetas matrizes,
- ids, vocabulário (id correspondência a palavra) e etiquetas.
Se você está montando os dados para um modelo de linguagem, você deve definir seus rótulos como sempre 0 a respeitar esses formatos. ## dados linguagem Modelo

Um modelo de linguagem é treinado para adivinhar o que a palavra seguinte está dentro de um fluxo de palavras. Não alimentá-lo as diferentes textos separadamente, mas concatenar-los todos juntos em uma grande variedade. Para criar os lotes, dividimos essa matriz em `pedaços bs` de textos contínuos. Note-se que em todas as tarefas PNL, que não utilizam a convenção usual de comprimento de sequência sendo a primeira dimensão de modo tamanho do lote é a primeira dimensão e comprimento da sequência é a segunda. Aqui você pode ler os pedaços de textos em linhas.

```python
path = untar_data(URLs.IMDB_SAMPLE)
data = TextLMDataBunch.from_csv(path, 'texts.csv')
x,y = next(iter(data.train_dl))
example = x[:15,:15].cpu()
texts = pd.DataFrame([data.train_ds.vocab.textify(l).split(' ') for l in example])
texts
```

```python
jekyll_warn("If you are used to another convention, beware! fastai always uses batch as a first dimension, even in NLP.")
```

Isso tudo é feito internamente quando usamos [`TextLMDataBunch`](/text.data.html#TextLMDataBunch), envolvendo o conjunto de dados na seguinte pré-loader antes de chamar um [`DataLoader`](https://pytorch.org/docs/stable/data.html#torch.utils.data.DataLoader).

```python
show_doc(LanguageModelPreLoader)
```

LanguageModelPreLoader é uma classe interna usa para treinar um modelo de linguagem. Leva as frases passados ​​como uma matriz denteada de frases em numericalised `dataset` e retorna lotes contíguos ao dataloader pytorch com o tamanho do lote bs`` e um comprimento de sequência `bptt`.
- `lengths` pode ser fornecido para os else comprimentos de dados de treino irregulares é calculado internamente
- `para trás = true` vai inverte as frases.
- `Shuffle = true`, irá embaralhar a ordem das frases, no início de cada época - exceto o primeiro
A descrição a seguir é útil para a compreensão da implementação de [`LanguageModelPreLoader`](/text.data.html#LanguageModelPreLoader):
- idx: exemplo de CircularIndex que indexa itens, tendo em conta o seguinte 1) baralhamento, 2) direcção da indexação, 3) envolve a cabeça (lendo para a frente) ou cauda (lendo para trás) da matriz irregular, conforme necessário, a fim de preencher o último lote (s)
- ro: índice do primeiro pano de cada linha no lote a ser extracto. Retorna como índice para o próximo de pano a ser extraído
- ri: Leitura para a frente: índice para o primeiro sinal a ser extraída no pano corrente (RO). Leitura para trás: uma posição após o último token a ser extraído no pano
- overlap: sobreposição entre lotes é 1, porque só prever o próximo token


## dados classificador

Ao preparar os dados para um classificador, mantemos os diferentes textos separados, o que representa outro desafio para a criação de lotes: uma vez que nem todos eles têm o mesmo comprimento, não podemos facilmente reunir-los juntos em lotes. Para ajudar com isso, usamos duas técnicas diferentes:
- padding: cada texto é preenchido com o `símbolo PAD` para obter todos os que escolheu para o mesmo tamanho
- classificando os textos (ish): para evitar ter juntos um texto muito longo com um muito curto (que teria, então, um monte de `fichas PAD`), nós reagrupar os textos por ordem de comprimento. Para o conjunto de treinamento, nós ainda adicionar um pouco de aleatoriedade para evitar mostrar os mesmos lotes em todas as etapas do treinamento.
Aqui é um exemplo de lote com estofamento (o índice de estofamento é um, e o enchimento é aplicado antes de iniciar as frases).
