A saída será `2 * sz`, ou apenas 2 se` sz` é Nenhum.

O objecto [`AdaptiveConcatPool2d`](/layers.html#AdaptiveConcatPool2d) usa o pool de média adaptativa e pooling max adaptativa e encadeia os dois. Nós usamos isso porque ele fornece o modelo com as informações de ambos os métodos e melhora o desempenho. Esta técnica é chamada `adaptive` porque nos permite decidir sobre o que as dimensões de saída que queremos, em vez de escolher as dimensões da entrada para caber um tamanho de saída desejado.
Vamos tentar o treinamento com Adaptive Média Pooling primeiro, depois com Adaptive Max Pooling e, finalmente, com a concatenação de ambos para ver como eles se saem no desempenho.
Vamos primeiro definir um [`simple_cnn`](/layers.html#simple_cnn) usando [Adaptive Max Pooling](https://pytorch.org/docs/stable/nn.html#torch.nn.AdaptiveMaxPool2d) alterando o código fonte um pouco.

````python
path = untar_data(URLs.MNIST_SAMPLE)
data = ImageDataBunch.from_folder(path)
````

````python
def simple_cnn_max(actns:Collection[int], kernel_szs:Collection[int]=None,
               strides:Collection[int]=None) -> nn.Sequential:
    "CNN with `conv2d_relu` layers defined by `actns`, `kernel_szs` and `strides`"
    nl = len(actns)-1
    kernel_szs = ifnone(kernel_szs, [3]*nl)
    strides    = ifnone(strides   , [2]*nl)
    layers = [conv_layer(actns[i], actns[i+1], kernel_szs[i], stride=strides[i])
        for i in range(len(strides))]
    layers.append(nn.Sequential(nn.AdaptiveMaxPool2d(1), Flatten()))
    return nn.Sequential(*layers)
````

````python
model = simple_cnn_max((3,16,16,2))
learner = Learner(data, model, metrics=[accuracy])
learner.fit(1)
````

Agora vamos tentar com [Adaptive Average Pooling](https://pytorch.org/docs/stable/nn.html#torch.nn.AdaptiveAvgPool2d) agora.

````python
def simple_cnn_avg(actns:Collection[int], kernel_szs:Collection[int]=None,
               strides:Collection[int]=None) -> nn.Sequential:
    "CNN with `conv2d_relu` layers defined by `actns`, `kernel_szs` and `strides`"
    nl = len(actns)-1
    kernel_szs = ifnone(kernel_szs, [3]*nl)
    strides    = ifnone(strides   , [2]*nl)
    layers = [conv_layer(actns[i], actns[i+1], kernel_szs[i], stride=strides[i])
        for i in range(len(strides))]
    layers.append(nn.Sequential(nn.AdaptiveAvgPool2d(1), Flatten()))
    return nn.Sequential(*layers)
````

````python
model = simple_cnn_avg((3,16,16,2))
learner = Learner(data, model, metrics=[accuracy])
learner.fit(1)
````

Finalmente, vamos tentar com a concatenação de ambos [`AdaptiveConcatPool2d`](/layers.html#AdaptiveConcatPool2d). Veremos que, na verdade, aumenta a nossa precisão e diminui nossa perda consideravelmente!

````python
def simple_cnn(actns:Collection[int], kernel_szs:Collection[int]=None,
               strides:Collection[int]=None) -> nn.Sequential:
    "CNN with `conv2d_relu` layers defined by `actns`, `kernel_szs` and `strides`"
    nl = len(actns)-1
    kernel_szs = ifnone(kernel_szs, [3]*nl)
    strides    = ifnone(strides   , [2]*nl)
    layers = [conv_layer(actns[i], actns[i+1], kernel_szs[i], stride=strides[i])
        for i in range(len(strides))]
    layers.append(nn.Sequential(AdaptiveConcatPool2d(1), Flatten()))
    return nn.Sequential(*layers)
````

````python
model = simple_cnn((3,16,16,2))
learner = Learner(data, model, metrics=[accuracy])
learner.fit(1)
````

Isto é muito útil para usar funções como camadas em nossas redes dentro de um objeto [Sequential](https://pytorch.org/docs/stable/nn.html#torch.nn.Sequential). Assim, por exemplo, dizer que queremos aplicar uma [log_softmax loss](https://pytorch.org/docs/stable/nn.html#torch.nn.functional.log_softmax) e precisamos mudar a forma dos nossos lotes de saída para ser capaz de usar essa perda. Podemos adicionar uma camada que se aplica a mudança necessária em forma chamando:
`Lambda (lambda x: x.view (x.size (0), - 1))`

Vamos ver um exemplo de como a forma da nossa produção pode mudar quando adicionamos essa camada.

````python
model = nn.Sequential(
    nn.Conv2d(3,  16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 10, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.AdaptiveAvgPool2d(1),
)

model.cuda()

for xb, yb in data.train_dl:
    out = (model(*[xb]))
    print(out.size())
    break
````

````python
model = nn.Sequential(
    nn.Conv2d(3,  16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 10, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.AdaptiveAvgPool2d(1),
    Lambda(lambda x: x.view(x.size(0),-1))
)

model.cuda()

for xb, yb in data.train_dl:
    out = (model(*[xb]))
    print(out.size())
    break
````

A função que nós construímos acima é realmente implementado na nossa biblioteca como [`Flatten`](/layers.html#Flatten). Podemos ver que ele retorna o mesmo tamanho quando executá-lo.

````python
model = nn.Sequential(
    nn.Conv2d(3,  16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 10, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.AdaptiveAvgPool2d(1),
    Flatten(),
)

model.cuda()

for xb, yb in data.train_dl:
    out = (model(*[xb]))
    print(out.size())
    break
````

Podemos combinar estas duas camadas finais ([AdaptiveAvgPool2d](https://pytorch.org/docs/stable/nn.html#torch.nn.AdaptiveAvgPool2d) e [`Flatten`](/layers.html#Flatten)) usando [`PoolFlatten`](/layers.html#PoolFlatten).

````python
model = nn.Sequential(
    nn.Conv2d(3,  16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 16, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    nn.Conv2d(16, 10, kernel_size=3, stride=2, padding=1), nn.ReLU(),
    PoolFlatten()
)

model.cuda()

for xb, yb in data.train_dl:
    out = (model(*[xb]))
    print(out.size())
    break
````

A função [`bn_drop_lin`](/layers.html#bn_drop_lin) retorna uma sequência de [batch normalization](https://arxiv.org/abs/1502.03167), [dropout](https://www.cs.toronto.edu/~hinton/absps/JMLRdropout.pdf) e uma camada linear. Esta camada personalizada é normalmente usado no final de um modelo.

````python
show_doc(conv_layer, doc_string=False)
````

A função [`conv_layer`](/layers.html#conv_layer) retorna uma sequência de [nn.Conv2D](https://pytorch.org/docs/stable/nn.html#torch.nn.Conv2d), [BatchNorm](https://arxiv.org/abs/1502.03167) e um Relu ou função de activação [leaky RELU](https://ai.stanford.edu/~amaas/papers/relu_hybrid_icml2013_final.pdf).
`N_in` representa o número de tamanho da entrada` n_out` o tamanho da saída, `tamanho do kernel ks`,` stride` o passo com o qual deseja aplicar as circunvoluções. `Bias` vai decidir se eles têm preconceito ou não (se nenhum, valor default true a menos que usando batchnorm). `Norm_type` selecciona o tipo de normalização (ou` none`). Se `leaky` é Nenhuma, a activação é um padrão` ReLU`, caso contrário, é um `` LeakyReLU` de inclinação leaky`. Finalmente, se transpor `= TRUE, a convolução é substituído por um` ConvTranspose2D`.
