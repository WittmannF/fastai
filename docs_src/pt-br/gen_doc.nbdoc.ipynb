## Mostrar a documentação de uma função

```python
show_doc(show_doc, doc_string=False)
```

Mostrar a documentação de um `elt` (função, classe ou enum). `Doc_string` decide se nós mostramos o doc string do elemento ou não,` full_name` irá substituir o nome mostrado, `arg_comments` é um dicionário que mostrará então lista os argumentos com comentários. `Title_level` é o nível da célula correspondente no índice,` alt_doc_string` é um texto que pode substituir a `doc_string`. `Ignore_warn` vai ignorar os avisos se você passar argumentos em` arg_comments` que não parecem pertencer a esta função e `markdown` decide se o retorno é uma célula Markdown ou texto simples.
A abundância de exemplos de usos dessa célula pode ser visto através da documentação, e você vai querer * hide entrada * essas células para um resultado final limpo.
