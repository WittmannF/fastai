[`PeakMemMetric`](/callbacks.mem.html#PeakMemMetric) é um retorno de chamada de memória de perfil.
Aqui está como você pode usá-lo:
```
from fastai.callbacks.mem import PeakMemMetric
learn = cnn_learner(data, model, metrics=[accuracy], callback_fns=PeakMemMetric)
learn.fit_one_cycle(3, max_lr=1e-2)
```
e um exemplo de saída:
```
Total time: 00:59
epoch	train_loss valid_loss accuracy cpu used peak gpu used peak
    1	0.325806   0.070334   0.978800	      0   2       80  6220
    2	0.093147   0.038905   0.987700	      0   2        2   914
    3	0.047818   0.027617   0.990600	      0   2        0   912
```
As últimas quatro colunas são o uso de memória deltas para CPU e GPU (em MBs).
* As colunas "de memória usada" mostrar a diferença entre o uso de memória antes e depois de cada época.
* As colunas "de memória de pico" Quanto sobrecarga de memória a época usados ​​em cima de memória utilizada. Com a rara exceção de medições GPU, onde se "memória usada" delta é negativo, então ele é calculado como uma diferença clara entre a memória de pico e a memória usada no início da época. Veja também
Por exemplo, na primeira linha da amostra de exemplo acima mostra `utilizada = 80`,` = pico 6220`. Isso significa que durante a execução deste segmento do aplicativo usado um máximo de 6300 MB ( `80 + 6220`), mas, em seguida, a maioria de que a memória foi lançado, mantendo apenas 80 MBs amarrados. Você pode então ver nas seguintes épocas que enquanto o aplicativo ainda usa memória temporária enquanto a execução, mas ele libera quase tudo no final do seu trabalho.
Além disso, é muito importante saber o alocador de memória desse pytorch pode trabalhar com menos memória, por isso, não significa que ele precisa de 6300 MB para ser capaz de executar a primeira época. Ele vai fazer com menos, mas ele vai ser apenas ligeiramente mais lento na primeira época. Para mais detalhes consulte [this explanation](dev/gpu.html#peak-memory-usage).
