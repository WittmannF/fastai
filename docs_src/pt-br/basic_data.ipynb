## Usando um conjunto de dados personalizado em fastai

Se você quiser usar o seu [`Dataset`](https://pytorch.org/docs/stable/data.html#torch.utils.data.Dataset) pytorch em fastai, você pode precisar de implementar mais atributos / métodos se você quiser usar a funcionalidade completa da biblioteca. Algumas funções pode ser facilmente utilizado com o seu [`Dataset`](https://pytorch.org/docs/stable/data.html#torch.utils.data.Dataset) pytorch se você acabou de adicionar um atributo, para outros, o melhor seria a de criar o seu próprio [`ItemList`](/data_block.html#ItemList) seguindo [this tutorial](/tutorial.itemlist.html). Aqui está uma lista completa do que a biblioteca vai esperar.

### Fundamentos

Primeiro de tudo, você obviamente precisa implementar os métodos `__len__` e` __getitem__`, como indicado pelas docs pytorch. Então as coisas mais necessárias seriam:
- atributo `c`: ele é usado na maioria das funções que criam diretamente um [`Learner`](/basic_train.html#Learner) ([`tabular_learner`](/tabular.learner.html#tabular_learner), [`text_classifier_learner`](/text.learner.html#text_classifier_learner), [`unet_learner`](/vision.learner.html#unet_learner), [`cnn_learner`](/vision.learner.html#cnn_learner)) e representa o número de saídas da camada final de seu modelo (também o número de classes se aplicável).
- atributo `classes`: é usado por [`ClassificationInterpretation`](/train.html#ClassificationInterpretation) e também em [`collab_learner`](/collab.html#collab_learner) (melhor usar [`CollabDataBunch.from_df`](/collab.html#CollabDataBunch.from_df) que um [`Dataset`](https://pytorch.org/docs/stable/data.html#torch.utils.data.Dataset) pytorch) e representa as marcas únicas que aparecem em seus dados.
- talvez um atributo `loss_func`: que vai ser usado por [`Learner`](/basic_train.html#Learner) como uma função perda padrão, por isso, se você sabe que seu [`Dataset`](https://pytorch.org/docs/stable/data.html#torch.utils.data.Dataset) personalizada requer uma perda particular, você pode colocá-lo.


exemplo brinquedo com arrays numpy imagem semelhante e etiqueta binário

```python
class ArrayDataset(Dataset):
    "Sample numpy array dataset"
    def __init__(self, x, y):
        self.x, self.y = x, y
        self.c = 2 # binary label
    
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, i):
        return self.x[i], self.y[i]
```

```python
train_x = np.random.rand(10, 3, 3) # 10 images (3x3)
train_y = np.random.rand(10, 1).round() # binary label

valid_x = np.random.rand(10, 3, 3)
valid_y = np.random.rand(10, 1).round()

train_ds, valid_ds = ArrayDataset(train_x, train_y), ArrayDataset(valid_x, valid_y)
data = DataBunch.create(train_ds, valid_ds, bs=2, num_workers=1)
data.one_batch()
```

Na tabela, o conjunto de dados terá de ter um atributo `cont_names` (para os nomes de variáveis ​​contínuas) e um método` get_emb_szs` que retorna uma lista de tuplas `(n_classes, emb_sz)` representando, para cada variável categórica, o número de códigos diferentes (não se esqueça de adicionar 1 para nan) eo tamanho incorporação correspondente. Aqueles dois são usados ​​com o atributo `c` por [`tabular_learner`](/tabular.learner.html#tabular_learner).
